## *Patch-Seq*
A technique that allows researchers to record an individual cell's electrical activity, inject dye into it for imaging, and also aspirate the cytoplasm and study the neuron's morphology.

The process of patch-seq involves patching a cell in a brain slice, recording its electrical activity, and then using suction to aspirate the cytoplasm and nucleus into the pipette for RNA extraction and sequencing.

It a powerful technique that allows many different measurements- electrical currents, function/shape, and gene expression. It allows researchers to study an isolated neuron.

Let's re-plot our histograms, and make the x-axis ranges the same. Let's also create x-axis and y-axis labels for our plots.

```python
# Let's re-make our histograms!

axes = cell_types['ISI'].hist(by = cell_types['Marker'])

axes[0].set_xlim(0,250)
axes[1].set_xlim(0,250)

# Enter appropriate labels for the x-axis and y-axis of the first histogram inside of the parentheses - remember to use '' when creating a string!

axes[0].set_ylabel('Frequency')
axes[0].set_xlabel('ISI (ms)')

# Create code that will do the same thing for your second histogram!

axes[1].set_ylabel('Frequency')
axes[1].set_xlabel('ISI (ms)')
```
