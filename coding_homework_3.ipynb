```python
#Imports Python libraries.

import numpy as np

import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
#Imports urls
url_sst = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
url_pvalb = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'

#Reads the data sets using numpy, and then saves them as v_rest_sst and v_rest_pvalb.
v_rest_sst = np.loadtxt(url_sst)
v_rest_pvalb = np.loadtxt(url_pvalb)
```

```python
#Finds the means of v_rest_sst and v_rest_pvalb
mean_sst = np.mean(v_rest_sst)
mean_pvalb = np.mean(v_rest_pvalb)
print(f'Mean of Sst resting membrane potentials: {mean_sst}')
print(f'Mean of Pvalb resting membrane potentials: {mean_pvalb}')

#Runs a t-test to test the null hypothesis that the means of v_rest_sst and v_rest_pvalb come from different distributions, and then print the results
stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
print(f'Statistic: {stat}, P-value: {p_val}')
if p_val < 0.05:
    print("We can REJECT the null hypothesis that the means of v_rest_sst and v_rest_pvalb come from different distributions.")
else:
    print("We can NOT reject the null hypothesis that the means of v_rest_sst and v_rest_pvalb come from different distributions.")

#Plots a histogram of v_rest_sst in one subplot, and a histogram of v_rest_pvalb in another subplot if the p-value of the t test is less that 0.05
if p_val < 0.05:
  fig, ax = plt.subplots(1,2)
  ax[0].hist(v_rest_sst)
  ax[1].hist(v_rest_pvalb)

#Labels the histogram's axes
ax[0].set_xlabel('Sst Resting Potentials')
ax[0].set_ylabel('Frequency')
ax[1].set_xlabel('Pvalb Resting Potentials')
ax[1].set_ylabel('Frequency')
```

```python
#Define a function that, when called upon, executes the code in the previous code cell
def vrestHist(x, y) :
  #Finds the means of x and y
  mean_x = np.mean(x)
  mean_y = np.mean(y)
  print(f'Mean of first dataset: {mean_x}')
  print(f'Mean of second dataset: {mean_y}')

  #Runs a t-test to test the null hypothesis that the means of x and y come from different distributions, and then print the results
  stat, p_val = stats.ttest_ind(x, y)
  print(f'Statistic: {stat}, P-value: {p_val}')
  if p_val < 0.05:
    print(f'We can REJECT the null hypothesis that the means of the two data sets come from different distributions.')
  else:
    print(f'We can NOT reject the null hypothesis that the means of of the two data sets come from different distributions.')

  #Plots a histogram of x in one subplot, and a histogram of y in another subplot if the p-value of the t test is less that 0.05
  if p_val < 0.05:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(x)
    ax[1].hist(y)

  #Labels the histogram's axes
  ax[0].set_xlabel('Sst Resting Potentials')
  ax[0].set_ylabel('Frequency')
  ax[1].set_xlabel('Pvalb Resting Potentials')
  ax[1].set_ylabel('Frequency')
```

```python
#Calls on the vrestHist() function defined above
vrestHist(v_rest_sst, v_rest_pvalb)
```
