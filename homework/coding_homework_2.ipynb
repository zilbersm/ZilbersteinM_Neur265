```python
#Imports python libraries
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
#Imports urls
url_sst = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
url_pvalb = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'

#Reads the data sets using numpy, and then saves them as v_rest_sst and v_rest_pvalb.
v_rest_sst = np.loadtxt(url_sst) #instruction 3 on GitHub says to name this "v_rest1" but instruction 4 assumes its named v_rest_sst...
v_rest_pvalb = np.loadtxt(url_pvalb) #instruction 3 on GitHub says to name this "v_rest2" but instruction 4 assumes its named v_rest_pvalb...
```

```python
#Concatenates v_rest_sst and v_rest_pvalb into one matrix.
v_rest = np.array((v_rest_sst,v_rest_pvalb))
```

```python
#Prints the size of v_rest
print(f'The size of v_rest is {v_rest.size}.')

#To confirm, find the sizes of the original two lists:
#print(f'The size of v_rest_sst is {v_rest_sst.size}, and the size of v_rest_pvalb is {v_rest_pvalb.size}.')
```

```python
#Prints the first row of v_rest
print(f'The first row of v_rest is {v_rest[0]}.')
```

```python
#Prints the sum of the values of the third column in v_rest
print(f'The sum of the values in column three is {v_rest[:,2].sum()}.')
```

```python
#Prints the second-to-last value in the second row of v_rest
print(f'The second-to-last value in the second row of v_rest is {v_rest[1, -2]}.')
```

```python
#Makes a histogram of the values in the first row of v_rest
plt.hist(v_rest[0])
plt.xlabel("Resting Membrane Potential (mV)")
plt.ylabel("Frequency")
```

```python
#Set variable v_rest_mean_sst equal to the mean of the first row, and set variable v_rest_mean_pvalb equal to the mean of the second row of v_rest.
v_rest_mean_sst = st.mean(v_rest[0])
v_rest_mean_pvalb = st.mean(v_rest[1])

#Print the means
print(f'The mean of the first row of v_rest is {v_rest_mean_sst}.')
print(f'The mean of the second row of v_rest is {v_rest_mean_pvalb}.')
```

##What is the difference between the two means?

From the code cell above, we found that v_rest_mean_sst = -67.73465864056 and v_rest_mean_pvalb = -69.97869017908.

So, we have v_rest_mean_sst > v_rest_mean_pvalb.

The resting membrane potential of Sst neurons are higher, on average, than the resting membrane potential of the Pvalb neurons.

```python
#Imports the scipy library
import scipy.stats as stats
```

```python
#Performs a t-test on the two rows of v_rest
stats.ttest_ind(v_rest[0], v_rest[1])
```



The results of the t-test indicate a statistically significant difference between the two groups (Sst vs. Pvalb neurons), with a t-statistic of $t(998) = 5.84$ and a $p$-value of $p < .001$ ($7.16 \times 10^{-9}$). Because the $p$-value is less than the standard threshold of $0.05$, we can conclude that the observed difference is unlikely to be due to chance.
