```python
# Stores link to the v_rest_sst data, then turned it into a Pandas Dataframe called pandas_data1.
url1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
pandas_data1 = pd.read_csv(url1)

# # Stores link to the v_rest_pvalb data, then turned it into a Pandas Dataframe called pandas_data2.
url2 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
pandas_data2 = pd.read_csv(url2)
```

```python
# Converts both Pandas DataFrames into lists.
data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
```

```python
# When Pandas Dataframes are converted to lists, each individual element is out in brackets.
# We can get around this by "flattening" our lists.
# According to google, in Python, flattening a list means converting a nested list structure (a list of lists) into a single, one-dimensional list, effectively removing the inner structure and placing all elements into a flat sequence.
#
# Flattens the lists made in the previous code block, thus getting rid of the brackets around each element.

data1 = np.squeeze(data1)
data2 = np.squeeze(data2)
```

```python
# Used the len() function to print the length of data1.

print(f' Length of list data1: {len(data1)}')
```

```python
# Printed the 3rd, 4th, 5th, and 6th elements of data2.

print(f'The 3rd, 4th, 5th, and 6th elements of list data2: {data2[2:6]}')
```

```python
# Printed the second-to-last and last elements of data1

print(f'The second-to-last and last elements of data1: {data1[-2:]}')
```

```python
# Printed the type of data of data2

print(f'The type of data in list data2 is {data2.dtype}')
```

```python
# Created a histogram of data1

plt.hist(data1)
```

```python
# Remade the histogram and labeled the axes.

plt.hist(data1)
plt.xlabel('Resting Membrane Potential (mV)')
plt.ylabel('Frequency')
```

```python
# Created a histogram with labeled axes for data2

plt.hist(data2)
plt.xlabel('Resting Membrane Potential (mV)')
plt.ylabel('Frequency')
```

## What did I notice?


*   Both histograms seem to peak around -70mV.
*   Both data sets have a similar range.
*   The histogram of data2 shows higher frequencies of member potential lower that -70mV in comparison to data1.
In other words, the pvalb ells seem to have slightly lower resting membrane potentials than the Sst cells!



```python
# Created a new set that contains both data1 and data2 (concatenated)

data3 = np.concatenate((data1,data2))
```
